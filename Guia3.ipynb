# Ejercicio 1
Generar una se√±al
- Cuadrada
- Senoidal
- Rampa
- Aleatoria

```python
Tini = 0    # Tiempo inicial
Tfin = 4    # Tiempo final
fm = 50     # Frecuencia de la muestra
T = 1/fm    # Periodo

t = np.arange(Tini,Tfin,T)
N = len(t)

# Onda cuadrada
y1 = [0]*N
for n in range(round(N/4)):
  y1[n] = 1
  y1[n+round(N/2)] = 1

# Senoidal
y2 = np.sin(2*np.pi*t)

# Rampa
y3 = np.mod(np.arange(0,N),fm)/fm

# Aleatoria
y4 = (np.random.random(size=N)*2)-1

# ------------------------ Plot ------------------------
plt.figure(figsize=(10,15))

plt.subplot(4,1,1)
plt.plot(t,y1)

plt.subplot(4,1,2)
plt.plot(t,y2)

plt.subplot(4,1,3)
plt.plot(t,y3)

plt.subplot(4,1,4)
plt.plot(t,y4)

plt.tight_layout()
plt.show()
```

y calcular los siguientes valores:
- Valor medio
- Maximo
- Minimo
- Amplitud
- Energia
- Accion
- Potencia Media
- Raiz del valor cuadratico medio

```python
# Funcion norma-p
def pnorm(x,p):
  if p == np.inf:
    return max(np.abs(x))

  N = len(x)
  return np.power(np.sum(np.power(np.abs(x),[p]*N),axis=0),1/p)
```

```python
print("CUADRADA")
print("valor medio y1: ", pnorm(y1,1)/len(y1))
print("maximo y1: ", max(y1))
print("minimo y1: ", min(y1))
print("amplitud y1: ", pnorm(y1,np.inf))
print("energia y1: ", pnorm(y1,2)**2)
print("accion y1: ", pnorm(y1,1))
print("potencia y1: ", np.sum(np.power(y1,[2]*N), axis=0)/N)
print("raiz del val cuadratico medio:", np.sqrt(np.sum(np.power(y1,[2]*N), axis=0)/N))
print()

print("SENOIDAL")
print("valor medio y1: ", pnorm(y2,1)/len(y2))
print("maximo y1: ", max(y2))
print("minimo y1: ", min(y2))
print("amplitud y1: ", pnorm(y2,np.inf))
print("energia y1: ", pnorm(y2,2)**2)
print("accion y1: ", pnorm(y2,1))
print("potencia y1: ", np.sum(np.power(y2,[2]*N), axis=0)/N)
print("raiz del val cuadratico medio:", np.sqrt(np.sum(np.power(y2,[2]*N), axis=0)/N))
print()

print("RAMPA")
print("valor medio y1: ", pnorm(y1,1)/len(y1))
print("maximo y1: ", max(y2))
print("minimo y1: ", min(y2))
print("amplitud y1: ", pnorm(y2,np.inf))
print("energia y1: ", pnorm(y2,2)**2)
print("accion y1: ", pnorm(y2,1))
print("potencia y1: ", np.sum(np.power(y2,[2]*N), axis=0)/N)
print("raiz del val cuadratico medio:", np.sqrt(np.sum(np.power(y2,[2]*N), axis=0)/N))
print()

print("ALEATORIA")
print("valor medio y1: ", pnorm(y2,1)/len(y2))
print("maximo y1: ", max(y2))
print("minimo y1: ", min(y2))
print("amplitud y1: ", pnorm(y2,np.inf))
print("energia y1: ", pnorm(y2,2)**2)
print("accion y1: ", pnorm(y2,1))
print("potencia y1: ", np.sum(np.power(y2,[2]*N), axis=0)/N)
print("raiz del val cuadratico medio:", np.sqrt(np.sum(np.power(y2,[2]*N), axis=0)/N))
print()
```

# Ejercicio 2

Este codigo **SOLO ES INTERACTIVO EN DESKTOP**.
Copiarlo y pegarlo en algun archivo .py para tener la version interactiva

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.widgets as pltw

def sin(Tini, Tfin, fs, A, phi, fm):
    Tm = 1/fm
    t = np.arange(Tini, Tfin-Tm, Tm)
    y = A * np.sin(2 * np.pi * fs * t + phi)
    return t, y

# Seno "canonico"
t,y = sin(0,2*np.pi,1,1,0,50)

# valores iniciales
fs_ini = 1      # Frecuencia de la onda
A_ini = 1       # Amplitud
phi_ini = 0     # Desplazamiento

# -------------------------- PLOT --------------------------
fig, ax = plt.subplots()

# seno "canonico" fijo
ax.plot(t,y)

# Seno que vamos a manipular
line, = ax.plot(t, sin(0,2*np.pi,fs_ini,A_ini,phi_ini,50)[1], lw=2)
ax.set_xlabel('t [s]')

# Hacer espacio para los Sliders
fig.subplots_adjust(left=0.25, bottom=0.3)

# Slider horizontal de frecuencia (fs)
axfs = fig.add_axes([0.25, 0.1, 0.65, 0.03])
fs_slider = pltw.Slider(
    ax=axfs,
    label='fs [Hz]',
    valmin=0.1,
    valmax=10,
    valinit=fs_ini,
)

# Slider horizontal de desplazamiento (phi)
axphi = fig.add_axes([0.25, 0.05, 0.65, 0.03])
phi_slider = pltw.Slider(
    ax=axphi,
    label='phi',
    valmin=0.1,
    valmax=2*np.pi,
    valinit=phi_ini,
)

# Slider vertical de amplitud (A)
axA = fig.add_axes([0.1, 0.25, 0.0225, 0.63])
A_slider = pltw.Slider(
    ax=axA,
    label="A",
    valmin=0,
    valmax=5,
    valinit=A_ini,
    orientation="vertical"
)

# texto con el valor del producto interno
text = fig.text(0.5, -0.25, '', transform=ax.transAxes, ha='center')
text.set_text('y1 ¬∑ y2 = {:.2f}'.format(np.vdot(y, sin(0,2*np.pi,fs_ini,A_ini,phi_ini,50)[1])))


# funcion a llamar cada vez que cambien los sliders
def update(val):
    yTest = sin(0,2*np.pi,fs_slider.val,A_slider.val,phi_slider.val,50)[1]
    line.set_ydata(yTest)
    text.set_text('y1 ¬∑ y2 = {:.2f}'.format(np.vdot(y, yTest)))
    fig.canvas.draw_idle()


# registrar el callback bajo el evento de on_changed
fs_slider.on_changed(update)
A_slider.on_changed(update)
phi_slider.on_changed(update)

# Boton para reiniciar los sliders
resetax = fig.add_axes([0.8, 0.025, 0.1, 0.04])
button = pltw.Button(resetax, 'Reset', hovercolor='0.975')


# Funcion de reset
def reset(event):
    fs_slider.reset()
    A_slider.reset()
    phi_slider.reset()
button.on_clicked(reset)


plt.show()
```

Grafica de producto interno en funcion en valores de
- A
- f
- fase phi

```python
fm = 50
T = 1/fm
t = np.arange(0,2*np.pi,T)

# senoidal de referencia
x = np.sin(2*np.pi*t)

# variar A [-0.5,+0.5)
dA = np.arange(0.5,1.5,0.01)
producto_dA = [0]*len(dA)
for n in range(len(dA)):
  producto_dA[n] = np.vdot(x,dA[n]*np.sin(2*np.pi*t))

# variar f [-0.5,+0.5]
df = np.arange(0.5,1.5,0.01)
producto_df = [0]*len(df)
for n in range(len(df)):
  producto_df[n] = np.vdot(x,np.sin(2*np.pi*df[n]*t))

# variar fase phi [0,pi/4]
dphi = np.arange(0,np.pi/4,0.01)
producto_dphi = [0]*len(dphi)
for n in range(len(dphi)):
  producto_dphi[n] = np.vdot(x,np.sin(2*np.pi*t+dphi[n]))


# -------------------- PLOT --------------------
plt.subplot(3,1,1)
plt.plot(dA,producto_dA)
plt.ylabel("<x,y>")
plt.xlabel("A")

plt.subplot(3,1,2)
plt.plot(df,producto_df)
plt.ylabel("<x,y>")
plt.xlabel("fs")

plt.subplot(3,1,3)
plt.plot(dphi,producto_dphi)
plt.ylabel("<x,y>")
plt.xlabel("phi")

plt.tight_layout()
plt.show()
```

# Ejercicio 3
Calcule el error cuadratico total en el ejemplo de aproximacion con funciones de Legendre

- con los coeficientes calculados en el ejemplo

```python
# definimos la aproximacion de lengendre a partir de 4 coeficientes
# en valores t
def eval_leg4(t,x0,x1,x2,x3):
  N = len(t)
  return x0/np.sqrt(2) + x1*np.sqrt(3/2)*t + x2*np.sqrt(5/2)*(3*np.power(t,[2]*N)/2 - 0.5) + x3*np.sqrt(7/2)*(5*np.power(t,[3]*N)/2 - 3*t/2)

def eval_leg6(t,x0,x1,x2,x3,x4,x5):
  N = len(t)
  first_4 = x0/np.sqrt(2) + x1*np.sqrt(3/2)*t + x2*np.sqrt(5/2)*(3*np.power(t,[2]*N)/2 - 0.5) + x3*np.sqrt(7/2)*(5*np.power(t,[3]*N)/2 - 3*t/2)
  return first_4 + x4*3*np.sqrt(2)*(35*np.power(t,[4]*N)-30*np.power(t,[2]*N)+3)/16 + x5*np.sqrt(11/2)*t*(63*np.power(t,[4]*N)-70*np.power(t,[2]*N)+15)/8
```

```python
# funcion a aproximar
fm = 50
T = 1/fm
t = np.arange(-1,1,T)
N = len(t)

y = [0]*N
for n in range(N):
  if t[n] < 0:
    y[n] = -1
  else:
    y[n] = 1

leg = eval_leg4(t,0,np.sqrt(3/2),0,-np.sqrt(7/32))

# ------------------------ PLOT ------------------------
plt.figure(figsize=(15,10))

plt.plot(t,y, label="y(t)")
plt.plot(t,leg, label="Legendre")
plt.legend()

plt.text(-1,1,f'Error Cuad Total: {pnorm(y-leg,2)**2}')

plt.tight_layout()
plt.show()
```

Calcular el error con
- peque√±as variaciones en los coeficientes de legendre. Construir una grafica 3D, con el error en z

```python
# variacion de los coeficientes
t1 = np.arange(-0.1,0.1,0.01)

alpha1 = np.sqrt(3/2)
alpha2 = -np.sqrt(7/32)

xaxis = []
yaxis = []
z = [0]*(len(t1)**2)

for i in range(len(t1)):
  xaxis += [t1[i]]*len(t1)
  yaxis = np.append(yaxis,t1)
  for j in range(len(t1)):
      auxsignal = [0]*len(t)
      auxsignal = eval_leg4(t,0,alpha1+t1[i],0,alpha2+t1[j])
      z[len(t1)*i + j] = pnorm(y-auxsignal,2)**2

# ------------------------ Plot ------------------------
fig = plt.figure(figsize=(9, 15))

ax = fig.add_subplot(111, projection='3d')
cmhot = plt.get_cmap("copper")
ax.scatter(xaxis,yaxis,z,vmin=min(z),vmax=max(z),c=z,cmap=cmhot)
ax.set_xlabel('a1')
ax.set_ylabel('a2')
ax.set_zlabel('<x,y>')

plt.tight_layout()
plt.show()
```

# Ejercicio 4
Generar una se√±al como CL del conjunto de se√±ales senoidales de frecuencia 1, 2, 3, 4, 5, 6, 7, 8, 9 y 10 Hz.
- Mida el grado de parecido con las se√±ales generadas representandolo en un grafico de barras

```python
# Genero las senoidales
fm = 200
T = 1/fm
t = np.arange(0,1,T)
N = len(t)

sins1 = [0]*10
for i in range(10):
  sins1[i] = np.sin(2*np.pi*(i+1)*t)

# Se√±al generada
a_CL = [.1,.3,.9,.2,.3,.7,.8,.5,.6,.1]

y = [0]*10
for i in range(10):
  y[i] = sins1[i]*a_CL[i]
y = np.sum(y,axis=0)

# Grado de parecido
prods = [0]*10
for i in range(10):
  prods[i] = np.vdot(sins1[i],y)

# --------------------- PLOT ---------------------

plt.subplot(4,1,1)
for i in range(10):
  plt.plot(t,sins1[i])

plt.subplot(4,1,2)
plt.plot(y)

plt.subplot(4,1,3)
plt.bar(range(10),prods,0.8)
plt.title("<x,y>")

plt.subplot(4,1,4)
plt.bar(range(10),a_CL,0.5)
plt.title("coeficientes")

plt.tight_layout()
plt.show()
```

- Vuelva a medir el parecido pero con una CL donde varia la fase de las senoidales

```python
fases = np.arange(0,np.pi,np.pi/10)

sins2 = [0]*10
for i in range(10):
  sins2[i] = np.sin(2*np.pi*t+fases[i])

# Se√±al generada
a_CL = [.1,.3,.9,.2,.3,.7,.8,.5,.6,.1]

y = [0]*10
for i in range(10):
  y[i] = sins2[i]*a_CL[i]
y = np.sum(y,axis=0)

# Grado de parecido
prods = [0]*10
for i in range(10):
  prods[i] = np.vdot(sins2[i],y)

# --------------------- PLOT ---------------------

plt.subplot(4,1,1)
for i in range(10):
  plt.plot(t,sins2[i])

plt.subplot(4,1,2)
plt.plot(t,y)

plt.subplot(4,1,3)
plt.bar(range(10),prods,0.8)
plt.title("<x,y>")

plt.subplot(4,1,4)
plt.bar(range(10),a_CL,0.5)
plt.title("coeficientes")

plt.tight_layout()
plt.show()
```

```python
# Cargar data
import scipy.io as loader
N,data = loader.wavfile.read('/content/escala.wav')
N = len(data)
print(data)

plt.figure(figsize=(80,10))
plt.plot(data)
plt.tight_layout()
plt.show()
```

```python
# Generar se√±ales de comparacion
fm = N/4
T = 1/fm
t = np.arange(0,4,T)
print("len data: ", len(data))
print("len t: ", len(t))

t_signals = np.arange(0,0.5,T)
print("len t_singals: ", len(t_signals))
notes = [0]*8
notes[0] = np.sin(2*np.pi*261.63*t_signals) # DO
notes[1] = np.sin(2*np.pi*293.66*t_signals) # RE
notes[2] = np.sin(2*np.pi*329.63*t_signals) # MI
notes[3] = np.sin(2*np.pi*349.23*t_signals) # FA
notes[4] = np.sin(2*np.pi*392.00*t_signals) # SOL
notes[5] = np.sin(2*np.pi*440.00*t_signals) # LA
notes[6] = np.sin(2*np.pi*493.88*t_signals) # SI
notes[7] = np.sin(2*np.pi*523.25*t_signals) # DO
N_signals = len(t_signals)
print("len notes: ", len(notes[0]))
print("N_signals: ", N_signals)

plt.figure(figsize=(15,10))

plt.subplot(8,1,1)
plt.plot(notes[0])
plt.title("DO")

plt.subplot(8,1,2)
plt.plot(notes[1])
plt.title("RE")

plt.subplot(8,1,3)
plt.plot(notes[2])
plt.title("MI")

plt.subplot(8,1,4)
plt.plot(notes[3])
plt.title("FA")

plt.subplot(8,1,5)
plt.plot(notes[4])
plt.title("SOL")

plt.subplot(8,1,6)
plt.plot(notes[5])
plt.title("LA")

plt.subplot(8,1,7)
plt.plot(notes[6])
plt.title("SI")

plt.subplot(8,1,8)
plt.plot(notes[7])
plt.title("DO")

plt.tight_layout()
plt.show()
```

```python
# Buscar LA
max_value = 0
max_i = 0
for i in range(8):
  print(f"Rango {i}: [{i*N_signals} : {(i+1)*N_signals}]")
  if (i+1)*N_signals > len(data):
    val = np.vdot(notes[5][:len(data)-i*N_signals],data[i*N_signals:])
  else:
    val = np.abs(np.vdot(notes[5],data[i*N_signals:(i+1)*N_signals]))
  print(f"producto interno {i}: ", val)
  if val > max_value:
    max_value = val
    max_i = i

print(f"LA se encuentra en el intervalo [{t[max_i*N_signals]},{t[(max_i+1)*N_signals]}]")
```
